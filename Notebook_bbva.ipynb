```python
# Utiliza groupby para agrupar por 'ID' y contar los valores únicos en 'period'
id_period_counts = balances.groupby('ID')['period'].nunique()

# Luego, puedes filtrar los ID que tienen más de un período
ids_with_multiple_periods = id_period_counts[id_period_counts > 1]

# Imprime los ID que tienen más de un período
print("ID con más de un período:")
print(ids_with_multiple_periods)
```
