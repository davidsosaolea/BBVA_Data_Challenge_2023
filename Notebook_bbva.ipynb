```python

# Pivotear el segundo DataFrame para obtener las columnas de balance y tarjeta por mes
df2_pivot = liabiities_1.pivot(index='ID', columns='month', values=['product_1', 'product_2']).fillna(0)
df2_pivot.columns = [f'{col[0]}_MES_{col[1]}' for col in df2_pivot.columns]

# Fusionar los DataFrames en base a la columna 'ID'
merged_df = pd.merge(train_1, df2_pivot, on='ID', how='left')

print(merged_df)
```
